```python
#Importing our dataset from csv file
import pandas as pd
dataset=pd.read_csv("/titanic-passengers (1).csv", sep=";")

dataset.head()
```

```python
#Preprocessing our data
def preprocess_data(new_data):
    new_data['Age'].fillna(new_data['Age'].mean(),inplace=True)
    new_data.replace({'Sex':{'male': 1,'female':0}},inplace=True)
    new_data['Cabin']=new_data.Cabin.fillna('G6')
    new_data.replace({'Survived':{'Yes': 1,'No':0}},inplace=True)
    return new_data
data=preprocess_data(dataset)
```

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
#features extraction
x=data[["Age","Sex", "Pclass", "Parch", "SibSp","Fare"]]
y= data["Survived"]

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)

#applying tree algorithm
dtree = tree.DecisionTreeClassifier()
dtree.fit(x_train, y_train)   #fitting our model
y_pred=dtree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
import graphviz
from sklearn import tree
dot_data = tree.export_graphviz(dtree, out_file=None, feature_names=x_train.columns, filled=True)
graph = graphviz.Source(dot_data)
graph.render("data")
graph

```
